```python
#import dependancies 
import pandas as pd
```

```python
#load and Read file

file_to_load = "vine_table.csv"

vine_review_df = pd.read_csv(file_to_load)
```

```python
#Filter the data to retrieve all the rows where the total_votes count is equal to or greater than 20 
total_votes_df = vine_review_df.loc[vine_review_df["total_votes"] >= 20]

total_votes_df
```

```python
#retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%
helpful_votes_df = total_votes_df.loc[(total_votes_df["helpful_votes"] / total_votes_df["total_votes"]) >= 0.5 ]


helpful_votes_df
```

```python
#retrieve all the rows where a review was written as part of the Vine program (paid)
paid_reviews_df = helpful_votes_df.loc[helpful_votes_df["vine"] == 'Y']

paid_reviews_df.head()
```

```python
# retrieve all the rows where the review was not part of the Vine program (unpaid)
unpaid_reviews_df = helpful_votes_df.loc[helpful_votes_df["vine"] == 'N']

unpaid_reviews_df.head()
```

```python
#determine total number of review paid
paid_reviews_df.info()

paid_count = paid_reviews_df["review_id"].count()

paid_count
paid_reviews_df.describe()
```

```python
#determine total number of review unpaid
unpaid_reviews_df.info()

unpaid_count = unpaid_reviews_df["review_id"].count()

unpaid_count
unpaid_reviews_df.describe()
```

```python
#determine number of 5 star reviews paid
paid_five_star_count = paid_reviews_df.loc[paid_reviews_df["star_rating"] >= 5].count()["review_id"]

paid_five_star_count
```

```python
#determine number of 5 star reviews unpaid
unpaid_five_star_count = unpaid_reviews_df.loc[unpaid_reviews_df["star_rating"] >= 5].count()["review_id"]

unpaid_five_star_count
```

```python
#percentage of 5 star review paid
percentage_paid = float(paid_five_star_count / paid_count) * 100

percentage_paid
```

```python
#percentage of 5 star review unpaid
percentage_unpaid = float(unpaid_five_star_count / unpaid_count) * 100

percentage_unpaid
```
